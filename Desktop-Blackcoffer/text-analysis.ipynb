<h3 style="color: red">=> Getting Blogs Content</h3>

```python
def extractingText(name,data):
    headingText = data.find('h1', {'class' : 'entry-title'}).get_text()    
    blogContent = data.find('div', {'class' : 'td-post-content'}).get_text()
    content = headingText + blogContent
    #print(content)
    try:
        file_read = open(f"./data/extract-text/{name}.txt","r",encoding="utf-8");
        file_text = file_read.read()
        file_read.close()
        print("File Readed")
    except:
        file = open(f"./data/extract-text/{name}.txt","w+",encoding="utf-8")
        file.write(content)
        file.close()
        print("File Created Successfully")
        
    print()
    print('-------------------------------------------------------------------------------------------')
    print()
```

<h3 style="color: red">=> Removing new Lines or \n from text</h3>

```python
input_files_txt = os.listdir('./data/extract-text/')
```

```python
special_char = list("~`!@#$%^&*()-_+=[]|\:;<>,/?—”“")
```

```python
def onlyText(obj):
    curr_file = open(f'./data/extract-text/{obj}','r',encoding="utf-8")
    curr_data = curr_file.read()
    curr_data = curr_data.replace('\n',' ')
    for wd in special_char:
        curr_data = curr_data.replace(wd,' ')
    curr_data = curr_data.replace('.','.')
    
    curr_data_list = curr_data.split(" ")
    curr_data_list_new = []
    for j in curr_data_list:
        if j != '':
            curr_data_list_new.append(j)
            
    return "  ".join(curr_data_list_new)
```

```python
def extractedRead():
    file = open(f"./data/filtered-text/{i}","r+",encoding="utf-8")
    file.read()
    file.close()
```

```python
for i in input_files_txt:
    print(i)
    print()
    fileter_content = onlyText(i)
    #print(fileter_content)
    
    try:
        extractedRead()
    except:
        file = open(f"./data/filtered-text/{i}","w+",encoding="utf-8")
        file.write(fileter_content)
        file.close()
        print("Filtered Successfully")
    
    print()
    print('------------------------------------------------------------------------------------------------------------')
    print()
```

# 14.) POLARITY SCORE CALCULATE

```python
polarity_score = []
for i in range(0,len(output_df)):
    var_positive_score = output_df.iloc[i]['Positive Score']
    var_negative_score = output_df.iloc[i]['Negative Score']
    var_polarity_score = (var_positive_score - var_negative_score) / ((var_positive_score + var_negative_score) + 0.000001)
    #print(f"Positive score => {var_positive_score} , Negative score => {var_negative_score} , Polarity Score => {round(var_polarity_score,2)}")
    polarity_score.append(var_polarity_score)
```

```python
output_df['Polarity Score'] = polarity_score
```

```python
output_df
```

# 15.) SUBJECTIVE SCORE CALCULATE

```python
subjective_score = []
for i in range(0,len(output_df)):
    var_positive_score = output_df.iloc[i]['Positive Score']
    var_negative_score = output_df.iloc[i]['Negative Score']
    var_total_words = output_df.iloc[i]['Word Count (Without Stop Word)']
    
    var_polarity_score = (var_positive_score + var_negative_score) / ((var_total_words) + 0.000001)
    #print(f"Positive score => {var_positive_score} , Negative score => {var_negative_score} , Polarity Score => {round(var_polarity_score,2)}")
    subjective_score.append(var_polarity_score)
```

# 16.) ANALYSIS OF READABILITY

```python
avg_sent_len_lst = []
percent_comp_words_lst = []
fog_ind_score = []

for i in range(0,len(output_df)):
    total_words = output_df.iloc[i]['Words Counts (All Words)']
    total_sentences = output_df.iloc[i]['Sentence Length']
    total_complex_words = output_df.iloc[i]['Complex Word Count']
    
    avg_sent_len = total_words / total_sentences
    per_comp_words = total_complex_words / total_words
    fog_ind = 0.4 * (avg_sent_len + per_comp_words)
    
    #print(f"Total Words => {total_words}, No of sentences => {total_sentences}, Total complex words => {total_complex_words}, Average sentence length => {round(avg_sent_len,2)}, Percentage of complex words => {round(per_comp_words,2)}, Fog Index => {round(fog_ind,2)}")
    
    avg_sent_len_lst.append(avg_sent_len)
    percent_comp_words_lst.append(per_comp_words)
    fog_ind_score.append(fog_ind)
    #print('-----------------------------------------------------------------------------------------------------------------------------')
```

# 18.) SYLLABLE PER WORD

```python
words_list = sentiment_df['Word'].values.tolist()
words_list
```

```python
def syllableCount(obj):
    total_syllable = 0
    for i in obj:
        i_upper = i.upper()
        if i_upper in words_list:
            syllable_val = sentiment_df[(sentiment_df['Word'] == i_upper)]['Syllables'].values.tolist()
            syllable= [str(j) for j in syllable_val]
            sylable_string = "".join(syllable)
            syllable_val_int = int(sylable_string)
            total_syllable += syllable_val_int
            #print("Total sylables available in file according to per word :- ", total_syllable)
    return total_syllable
```

# 19.) AVERAGE WORD LENGTH

```python
def wordChars(obj):
    total_chars = []
    for i in obj:
        if len(i) > 1:
            for j in i:
                if j != ".":
                    total_chars.append(j)                
    total_chars = len(total_chars)
    return total_chars
```
