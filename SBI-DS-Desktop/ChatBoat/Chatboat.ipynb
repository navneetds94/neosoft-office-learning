```python
import nltk
import numpy as np
import random
import string
import requests
from bs4 import BeautifulSoup
import re
import pandas as pd
```

```python
content = "https://en.wikipedia.org/wiki/Chatbot"
```

```python
headers = {'User-Agent': "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/42.0.2311.135 Safari/537.36 Edge/12.246"}
# Here the user agent is for Edge browser on windows 10. You can find your browser user agent from the above given link.
raw_data = requests.get(url=content, headers=headers)
print(raw_data.content)
```

```python
soup = BeautifulSoup(raw_data.content, 'html5lib') # If this line causes an error, run 'pip install html5lib' or install html5lib
print(soup.prettify())
```

```python
need_html = soup.find('div', attrs = {'id':'mw-content-text'})
need_html
```

```python
need_html = need_html.find('div',attrs={'class':'mw-parser-output'})
need_html
```

# STORING TEXT DATA IN FILE

```python
count = 0
texture = ""
for tag_row in need_html:
    print(f"########################## Tag Row Number {count} ##########################")
    content = tag_row.get_text()
    print(content)
    texture += content
    
fs = open('./web-scrapping-text.txt','w+',encoding="utf-8")
fs.write(texture)
fs.close()
count+=1
```

# READ WEBSCRAPING FILE DATA

```python
fs = open('./web-scrapping-text.txt','r+',encoding="utf-8")
corpus = fs.read()
fs.close()
```

```python
corpus
```

# BLANK DATAFRAME

```python
df = pd.DataFrame()
```

# ADDING CORPUS TEXT IN DATAFRAME

```python
df['Text'] = [corpus]
df['Text']
```

```python
df['Text'][0]
```

```python
df
```

```python
def unwantedText(ind):
    remove_new_line = re.sub('\s',' ',df['Text'][ind])
    return re.sub('[^a-zA-Z0-9.()]',' ',remove_new_line)
```

```python
df_rows = len(df['Text'])
df['Corpus'] = [unwantedText(x) for x in range(0, df_rows)]
```

```python
df['Corpus Lower'] = df['Corpus'].apply(lambda x: x.lower())
```

```python
df
```

```python
def onlyText(ind):
    return re.sub('[^a-zA-Z]',' ',df['Corpus Lower'][ind])
```

```python
df["Corpus Text"] = [onlyText(x) for x in range(0, df_rows)]
```

```python
df
```

```python
from nltk.tokenize import word_tokenize , sent_tokenize
```

```python
df['Words'] = df['Corpus Text'].apply(lambda x: word_tokenize(x))
```

```python
import nltk
from nltk.corpus import stopwords
```

```python
english_stopwords = stopwords.words('english')
english_stopwords
```

```python
# stringList = df['Words'][0]
# stringList
```

```python
def removeStopWords(stringList):
    return [x for x in stringList if x not in english_stopwords]
```

```python
check1 = removeStopWords(df['Words'][0])
check1
```

```python
df['Without Stopwords'] = df['Words'].apply(lambda x: removeStopWords(x))
```

```python
df
```

```python
from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer
```

```python
lm = WordNetLemmatizer()
```

```python
ps = PorterStemmer()
```

```python
def wordPorterStemmer(val):
    return [ps.stem(val[x]) for x in range(0,len(val))]
```

```python
def wordLemmitizer(val):
    return [lm.lemmatize(val[x],pos="v") for x in range(0,len(val))]
```

```python
lem_list = df['Without Stopwords'][0]
lem_list
```

```python
lm.lemmatize("simulates")
```

```python
check1 = wordLemmitizer(lem_list)
check1
```

```python
check2 = wordPorterStemmer(lem_list)
check2
```

```python
df['Lemmentise'] = df['Without Stopwords'].apply(lambda x: wordLemmitizer(x))
df['Porter'] = df['Without Stopwords'].apply(lambda x: wordPorterStemmer(x))
```
