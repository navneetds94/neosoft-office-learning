```python
import pandas as pd
```

```python
gdp = pd.read_csv('./gdp-capital.csv')
gdp.head(5)
```

```python
def appendIMF(s,c,r,imfa,imfy):
  bank_name = "IMF"
  # print(bank_name)
  # print(s)
  # print(c)
  # print(r)
  # print(imfa)
  # print(imfy)
  # print("-------------------- Seprator --------------------")
  gdp_formated_dict["Sno"].append(s)
  gdp_formated_dict["Country/Territory"].append(c)
  gdp_formated_dict["UN Region"].append(r)
  gdp_formated_dict["Bank"].append(bank_name)
  gdp_formated_dict["Ammount"].append(imfa)
  gdp_formated_dict["Year"].append(imfy)
```

```python
def appendWorldBank(s,c,r,wba,wby):
  bank_name = "World Bank"
  # print(bank_name)
  # print(s)
  # print(c)
  # print(r)
  # print(wba)
  # print(wby)
  # print("-------------------- Seprator --------------------")
  gdp_formated_dict["Sno"].append(s)
  gdp_formated_dict["Country/Territory"].append(c)
  gdp_formated_dict["UN Region"].append(r)
  gdp_formated_dict["Bank"].append(bank_name)
  gdp_formated_dict["Ammount"].append(wba)
  gdp_formated_dict["Year"].append(wby)
```

```python
def appendUnitedNations(s,c,r,una,uny):
  bank_name = "United Nations"
  # print(bank_name)
  # print(s)
  # print(c)
  # print(r)
  # print(una)
  # print(uny)
  # print("-------------------- Seprator --------------------")
  gdp_formated_dict["Sno"].append(s)
  gdp_formated_dict["Country/Territory"].append(c)
  gdp_formated_dict["UN Region"].append(r)
  gdp_formated_dict["Bank"].append(bank_name)
  gdp_formated_dict["Ammount"].append(una)
  gdp_formated_dict["Year"].append(uny)
```

```python
gdp_formated_dict = {"Sno":[],"Country/Territory":[],"UN Region":[],"Bank":[],"Ammount":[],"Year":[]}
gdp_formated_dict

for i in range(0,len(gdp)):
  print(f"Data {i+1} load at this iloc {i}")
  sno = gdp['Sno'].iloc[i]
  country = gdp['Country/Territory'].iloc[i]
  region = gdp['UN Region'].iloc[i]
  imfAmount = gdp['IMF'].iloc[i]
  imfYear = gdp['IMF-Year'].iloc[i]
  worldBankAmount = gdp['World Bank'].iloc[i]
  worldBankYear = gdp['World Bank - Year'].iloc[i]
  unitedNationsAmount = gdp['United Nations'].iloc[i]
  unitedNationsYear = gdp['World Bank - Year'].iloc[i]

  # IMF Change
  appendIMF(sno,country,region,imfAmount,imfYear)

  # World Bank Change
  appendWorldBank(sno,country,region,worldBankAmount,worldBankYear)

  # United Nations Change
  appendUnitedNations(sno,country,region,unitedNationsAmount,unitedNationsYear)
  
  print()
  print('----------------------------------------------------------------------')
  print()
```
