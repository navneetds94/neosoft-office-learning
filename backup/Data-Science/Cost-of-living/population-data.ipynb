```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
country_population = {}
state_population = {}
state_population = {}
pop_keys = []
```

```python
T = True
F = False
```

```python
def switchCase(index):
        switcher = {
            0: "#",
            1: "Country (or dependency)",
            2: "Population (2020)",
            3: "Yearly Change",
            4: "Net Change",
            5: "Density (P/Km²)",
            6: "Land Area (Km²)",
            7: "Migrants (net)",
            8: "Fert. Rate",
            9: "Med. Age",
            10: "Urban Pop %",
            11: "World Share"
        }
        if index in switcher:
            return switcher[index]
        else:
            return "Not Found"
```

```python
# Making a GET request
country_request = requests.get('https://www.worldometers.info/world-population/population-by-country/')
 
# Parsing the HTML
soup = BeautifulSoup(country_request.content, 'html5lib')
 
table_data = soup.find('table', attrs={'id':"example2"})
content = table_data
```

```python
content_headings = content.find_all('th')
print('table headings');
for heading in content_headings:
    print(heading.text, end=", ")
    pop_keys.append(heading.text)
    
for i in pop_keys:
    country_population[i] = []
```

```python
state_keys = []
def getStateData(link):
    try:
        state_request = requests.get(link)
        state_full = BeautifulSoup(state_request.content, 'html5lib')
        state_table = state_full.find('table',attrs={'class':"table table-hover table-condensed table-list"})
        state_heading = state_table.find_all('th')
        state_keys.append('Country')
        
        for th in state_heading:

            if th.text not in state_keys:
                state_keys.append(th.text)
                return state_keys
            else:
                print(state_keys)
                return 'Already Available'
                continue
    except:
        print('Not Found')
        pass
    
```

```python
check1 = getStateData('https://www.worldometers.info/world-population/mexico-population/')
check1
```

```python
content_rows_list = content.find_all('tr')[1:]
for row in range(len(content_rows_list)):
    td_data_items = content_rows_list[row].find_all('td')
    for td_text in range(len(td_data_items)):
        if switchCase(td_text) != "Not Found":
            print(f'{switchCase(td_text)} => {td_data_items[td_text].text},')
            country_population[switchCase(td_text)].append(td_data_items[td_text].text)
            if td_text == 1:
                city_name = td_data_items[td_text].text.lower()
                a_tag = td_data_items[td_text].find_all('a')[0]['href']
                print(f'https://www.worldometers.info{a_tag}')
                #print(getStateData(f'https://www.worldometers.info{a_tag}'))
        else:
            break;
    #break;
```

```python
population_df = pd.DataFrame(country_population)
population_df.head()
```

```python
population_df.set_index('#',inplace=T)
population_df = population_df.sort_values('Country (or dependency)').reset_index().drop('#',axis=1)
```

```python
population_df.replace('N.A.','0',inplace=T)
population_df.replace(' ','0',inplace=T)
```

```python
population_df['Population (2020)'] = population_df['Population (2020)'].apply(lambda x: int(x.replace(',','')))
population_df['Population (2020)'].head(5)
```

```python
population_df['Yearly Change'] = population_df['Yearly Change'].apply(lambda x: float(x.replace(' %','')))
population_df['Yearly Change'].head(5)
```

```python
population_df['Density (P/Km²)'] = population_df['Density (P/Km²)'].apply(lambda x: int(x.replace(',','')))
population_df['Density (P/Km²)'].head(5)
```

```python
population_df['Land Area (Km²)'] = population_df['Land Area (Km²)'].apply(lambda x: int(x.replace(',','')))
population_df['Land Area (Km²)'].head(5)
```

```python
population_df['Migrants (net)'] = population_df['Migrants (net)'].apply(lambda x: int(x.replace(',','')))
population_df['Migrants (net)'].head(5)
```

```python
population_df['Fert. Rate'] = population_df['Fert. Rate'].astype('float32')
population_df['Fert. Rate'].head(5)
```

```python
population_df['Med. Age'] = population_df['Med. Age'].astype('int32')
population_df['Med. Age'].head(5)
```

```python
population_df['Urban Pop %'] = population_df['Urban Pop %'].apply(lambda x: float(x.replace(' %','')))
population_df['Urban Pop %'].head(5)
```

```python
population_df['World Share'] = population_df['World Share'].apply(lambda x: float(x.replace(' %','')))
population_df['World Share'].head(5)
```
